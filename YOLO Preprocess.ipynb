```
import cv2
import glob
from faced import FaceDetector
from faced.utils import annotate_image
```

## Preprocessing

```
face_detector = FaceDetector()
```

```
!mkdir faces
```

```
for filename in glob.glob("*.mp4"):
    cap = cv2.VideoCapture(filename)
    iter = 0
    while True:
      has_frame, frame = cap.read()
      if not has_frame:
        break

      # Receives RGB numpy image (HxWxC) and
      # returns (x_center, y_center, width, height, prob) tuples. 
      bboxes = face_detector.predict(frame)
      for bbox in bboxes:
        startX = int(bbox[0]-(bbox[2]/2))
        endX = int(bbox[0]+(bbox[2]/2))
        startY = int(bbox[1]-(bbox[3]/2))
        endY = int(bbox[1]+(bbox[3]/2))
        image = frame[startY:endY, startX:endX]
        cv2.imwrite('faces/' + str(filename[:-4]) + '_' + str(iter)+".jpg", image)
        iter+=1

        print(bboxes)
```
